```python
from openai import OpenAI
import os
from dotenv import load_dotenv
from IPython.display import display, Markdown
load_dotenv(os.path.join("..", ".env"), override=True)
```

```python
dsp_prompt_1 = """
Act as a customer service analyst. Your task is to generate a structured report from the provided email.

[Original Text]:
"Hi, my order #4459 arrived yesterday but the box was completely crushed. Also, the delivery guy from the third-party company was very rude when I asked him to let me check the package. To make matters worse, I tried to use the discount code 'SUMMER20' on your website for a future purchase and the page threw a 500 error. I am very upset, I demand a refund for the shipping cost."

[Directional Stimulus / Hints]:
- Logistics issue (package condition and delivery attitude).
- Technical issue (website failure with coupon).
- Final demand (shipping refund).

Instruction: Based on the original text and strictly guided by the Directional Stimulus, write a brief three-bullet report routing each issue to its corresponding department.
"""
```

```python
client = OpenAI()

def run_prompt(prompt):
    response = client.chat.completions.create(
        model="gpt-4o",
        messages=[
            {"role": "user", "content": prompt}
        ],
        temperature=0.3
    )

    display(Markdown(response.choices[0].message.content))
```

```python
run_prompt(dsp_prompt_1)
```
